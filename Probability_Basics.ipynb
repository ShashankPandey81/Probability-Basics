is the standard error of the mean (SEM)?

- Population standard deviation,
σ=₹5,000

  Sample size, n=100

  Standard Error of Mean (SEM) = σ / √N

  SEM = 5000 / √100 = 5000 / 10 = 500

  This means the sample mean is expected to vary by about ₹500 from the population mean

Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that
both are Aces?

- Total Number of Cards = 52
  
-  Total Aces = 4

 - Cards are drawn without replacement

 Probability of first card is an Ace is = 4/52 = **1/13**

 Probability second card is an Ace = 3/52 = **1/17**

 Probability of both aces = 1/13 X 1/17
  
 P(both aces) = 1/221

 So, the chance of drawing two Aces in a row is 1 out of 221.

Question 10: Differentiate between discrete and continuous random variables with examples.

- Discrete takes countable values like whole number ( 1,2,3,4,5,6,7) and it has probability for each exact value .

Examples of Discrete -

Number of heads when tossing 3 coins → {0, 1, 2, 3}

Number of goals scored in a match

- Continuous Random variable takes uncountable values Any value within a range and Probability of Exact value is 0 .

Examples of Continuous -

Height of a student (e.g., 160.2 cm, 160.25 cm)

Time taken to complete a task
